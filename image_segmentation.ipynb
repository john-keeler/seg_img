```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# <font color=blue>Algoritmo de Threshold</font>
El algoritmo de Threshold es uno de los algoritmos introductorios en la segmentación de imágenes. Este caso en específico realiza una segmentación binaria de una imagen en escala de grises. 
<br>
Básicamente solo consta de una función pero dicha función queda en base a 4 parámetros de los cuales para este caso 3 son fijos pues se tratan del nombre de la imagen y el tipo de segmentación.
<br>
<font color=blue>retval, dst</font>	=	cv.threshold(	src, thresh, maxval, type[, dst]	)
### Parámetros

    src   input array (multiple-channel, 8-bit or 32-bit floating point).
    thresh   threshold value.
    maxval   maximum value to use with the THRESH_BINARY and THRESH_BINARY_INV thresholding types.
    type   thresholding type (see the cv::ThresholdTypes).

```python
img = cv2.imread('puente.jpg',0)
# img = cv2.medianBlur(img,5)

ret, th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret, th2 = cv2.threshold(img,150,255,cv2.THRESH_BINARY)
ret, th3 = cv2.threshold(img,160,255,cv2.THRESH_BINARY)
ret, th4 = cv2.threshold(img,170,255,cv2.THRESH_BINARY)

ret, th5 = cv2.threshold(img,127,245,cv2.THRESH_BINARY)
ret, th6 = cv2.threshold(img,150,235,cv2.THRESH_BINARY)
ret, th7 = cv2.threshold(img,160,225,cv2.THRESH_BINARY)
ret, th8 = cv2.threshold(img,170,215,cv2.THRESH_BINARY)

titulos = ['Imagen original', 'Parametros 127-255', 'Parametros 150-250',
            'Parametros 160-255', 'Parametros 170-255', 'Parametros 127-245',
            'Parametros 150-235', 'Parametros 160-225', 'Parametros 170-215']
imagenes = [img, th1, th2, th3, th4, th5, th6, th7, th8]

for i in range(9):
    plt.figure(figsize=(40,40))
    plt.subplot(9,1,i+1),plt.imshow(imagenes[i],'gray')
    plt.title(titulos[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
